```python
import gdown
import pandas as pd

# ID file dari link Google Drive
file_id = '10oJeOxJrW3FhkxL_Ee6C0ftTv99GOFsg'
url = f'https://drive.google.com/uc?id={file_id}'

# Download file menggunakan gdown
output = 'your_file.csv'
gdown.download(url, output, quiet=False)

# Baca CSV menggunakan pandas
df = pd.read_csv(output)
print(df.head())
```

```python
missing_values = df.isnull().sum()
```

```python
df_cleaned = df.dropna()
```

```python
print(df.dtypes)
```

```python
# Convert 'ProductPrice' to numeric (float), removing any non-numeric characters
df['ProductPrice'] = pd.to_numeric(df['ProductPrice'], errors='coerce')

# Convert 'OrderDate' to datetime format
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
df.head(5)
```

```python
# Menghitung Revenue, Expenses, dan Profit
df['Revenue'] = df['OrderQuantity'] * df['ProductPrice']
df['Expenses'] = df['OrderQuantity'] * df['ProductCost']
df['Profit'] = df['Revenue'] - df['Expenses']

# Tampilkan beberapa baris data
df[['OrderID', 'Revenue', 'Expenses', 'Profit']].head()
```

```python
# Menghitung ROI dan Profit Margin
df['ROI'] = (df['Profit'] / df['Expenses']) * 100
df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100

# Tampilkan hasil analisis kinerja
df[['OrderID', 'ROI', 'Profit_Margin']].head()
```

```python
# Mengelompokkan data berdasarkan bulan dan menganalisis tren pendapatan dan pengeluaran
monthly_trends = df.groupby(df['OrderDate'].dt.to_period('M')).agg(
    total_revenue=('Revenue', 'sum'),
    total_expenses=('Expenses', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

# Tampilkan tren keuangan bulanan
monthly_trends.head()
```

```python
import matplotlib.pyplot as plt

# Membuat plot tren pendapatan, pengeluaran, dan keuntungan dengan warna pastel
plt.figure(figsize=(10, 6))

# Warna pastel untuk garis
plt.plot(monthly_trends['OrderDate'].astype(str), monthly_trends['total_revenue'], label='Total Revenue', color='#AEC6CF', linewidth=2)  # Pastel Blue
plt.plot(monthly_trends['OrderDate'].astype(str), monthly_trends['total_expenses'], label='Total Expenses', color='#FFB6C1', linewidth=2)  # Pastel Pink
plt.plot(monthly_trends['OrderDate'].astype(str), monthly_trends['total_profit'], label='Total Profit', color='#B2DFEE', linewidth=2)  # Pastel Cyan

plt.title('Monthly Financial Trends')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Menyimpan grafik tren dengan warna pastel
plt.savefig('monthly_financial_trends_pastel.png')
```

```python
# Memfilter transaksi yang merupakan fraud atau kegagalan pembayaran
failed_transactions = df[df['Fraud'] == 1]  # Asumsikan nilai 1 untuk Fraud

# Tampilkan transaksi yang gagal
failed_transactions[['OrderID', 'TransactionID', 'PaymentMethod', 'Fraud']].head()
```

```python
# Mengelompokkan kegagalan pembayaran berdasarkan PaymentMethod
payment_failure_reasons = failed_transactions['PaymentMethod'].value_counts()

# Tampilkan hasil analisis penyebab kegagalan pembayaran
payment_failure_reasons.head()
```

```python
# Menghitung total kerugian finansial dari transaksi yang gagal
failed_transactions.loc[:, 'FinancialImpact'] = failed_transactions['OrderQuantity'] * failed_transactions['ProductCost']

# Total dampak finansial dari kegagalan pembayaran
total_loss = failed_transactions['FinancialImpact'].sum()

print(f"Total Financial Impact from Failed Transactions: ${total_loss}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menghitung alasan kegagalan pembayaran
payment_failure_reasons = df_cleaned['PaymentMethod'].value_counts()

# Mendapatkan index dari nilai maksimum dan minimum
max_index = np.argmax(payment_failure_reasons)
min_index = np.argmin(payment_failure_reasons)

# Menyiapkan warna: magenta untuk maksimum, pastel pink untuk minimum, dan warna default untuk lainnya
colors = ['#FFB6C1'] * len(payment_failure_reasons)  # Pastel pink untuk semua bagian
colors[max_index] = 'magenta'  # Magenta untuk nilai maksimum

# Membuat pie chart
plt.figure(figsize=(7, 5))
plt.pie(payment_failure_reasons, labels=payment_failure_reasons.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Most Common Reasons for Payment Failures')
plt.tight_layout()
plt.show()

# Menyimpan grafik
plt.savefig('payment_failure_reasons_pie_chart.png')
```

```python
# Mengelompokkan data berdasarkan hari, minggu, dan bulan
daily_volume = df.groupby(df['OrderDate'].dt.date).agg(daily_transactions=('OrderID', 'count')).reset_index()
weekly_volume = df.groupby(df['OrderDate'].dt.to_period('W')).agg(weekly_transactions=('OrderID', 'count')).reset_index()
monthly_volume = df.groupby(df['OrderDate'].dt.to_period('M')).agg(monthly_transactions=('OrderID', 'count')).reset_index()

# Tampilkan beberapa data volume transaksi harian
daily_volume.head()
```
