```python
import gdown
import pandas as pd

# ID file dari link Google Drive
file_id = '10oJeOxJrW3FhkxL_Ee6C0ftTv99GOFsg'
url = f'https://drive.google.com/uc?id={file_id}'

# Download file menggunakan gdown
output = 'your_file.csv'
gdown.download(url, output, quiet=False)

# Baca CSV menggunakan pandas
df = pd.read_csv(output)
print(df.head())
```

```python
# Convert 'ProductPrice' to numeric (float), removing any non-numeric characters
df['ProductPrice'] = pd.to_numeric(df['ProductPrice'], errors='coerce')

# Convert 'OrderDate' to datetime format
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
```

```python
# Menghitung total sales untuk setiap produk
# Asumsikan 'OrderQuantity' adalah jumlah yang dipesan dan 'ProductPrice' adalah harga produk

df['TotalSales'] = df['OrderQuantity'] * df['ProductPrice']  # Menghitung total sales untuk setiap baris

# Mengelompokkan berdasarkan ProductSKU dan menghitung total sales untuk setiap produk
best_sellers = df.groupby('ProductSKU').agg(total_sales=('TotalSales', 'sum')).reset_index()

# Mengurutkan produk dari penjualan tertinggi ke terendah
best_sellers_sorted = best_sellers.sort_values(by='total_sales', ascending=False)

# Menampilkan produk terlaris
best_sellers_sorted.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Memilih top 10 produk dengan penjualan tertinggi
top_10_products = best_sellers_sorted.head(10)

# Mencari index bar dengan nilai total sales tertinggi
max_sales_index = np.argmax(top_10_products['total_sales'])

# Menyiapkan warna: Semua bar pastel pink kecuali yang tertinggi magenta
colors = ['#FFB6C1'] * len(top_10_products)  # Pastel pink untuk semua bar
colors[max_sales_index] = 'magenta'  # Magenta untuk bar dengan nilai tertinggi

# Membuat bar chart untuk top 10 produk
plt.figure(figsize=(10, 6))
plt.bar(top_10_products['ProductSKU'], top_10_products['total_sales'], color=colors)
plt.ylabel('Total Sales')
plt.title('Top 10 Best-Selling Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menyimpan grafik
plt.savefig('top_10_best_selling_products.png')
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Mengambil data terbaru yang telah di-load sebelumnya
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Filter untuk data dalam seminggu terakhir dari tanggal terakhir di dataset
last_date = df['OrderDate'].max()
one_week_ago = last_date - timedelta(days=7)

# Filter data untuk penjualan dalam seminggu terakhir
df_last_week = df[df['OrderDate'] >= one_week_ago]

# Menghitung total sales untuk setiap produk dalam seminggu terakhir
df_last_week['TotalSales'] = df_last_week['OrderQuantity'] * df_last_week['ProductPrice']

# Mengelompokkan berdasarkan ProductSKU dan menghitung total sales untuk setiap produk
trending_products = df_last_week.groupby('ProductSKU').agg(total_sales=('TotalSales', 'sum')).reset_index()

# Mengurutkan produk dari penjualan tertinggi ke terendah
trending_products_sorted = trending_products.sort_values(by='total_sales', ascending=False)

# Menampilkan produk yang sedang tren
trending_products_sorted.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Memilih top 10 produk yang sedang tren
top_10_trending_products = trending_products_sorted.head(10)

# Mencari index bar dengan nilai total sales tertinggi
max_trending_sales_index = np.argmax(top_10_trending_products['total_sales'])

# Menyiapkan warna: Semua bar pastel pink kecuali yang tertinggi magenta
colors = ['#FFB6C1'] * len(top_10_trending_products)  # Pastel pink untuk semua bar
colors[max_trending_sales_index] = 'magenta'  # Magenta untuk bar dengan nilai tertinggi

# Membuat bar chart untuk top 10 produk yang sedang tren
plt.figure(figsize=(10, 6))
plt.bar(top_10_trending_products['ProductSKU'], top_10_trending_products['total_sales'], color=colors)
plt.ylabel('Total Sales in the Last Week')
plt.title('Top 10 Trending Products in the Last Week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menyimpan grafik
plt.savefig('top_10_trending_products_last_week.png')
```

```python
# Menghitung total pengeluaran (Total Sales) dan jumlah kunjungan untuk setiap pelanggan
df['TotalSales'] = df['OrderQuantity'] * df['ProductPrice']

customer_data = df.groupby('CustID').agg(
    total_spent=('TotalSales', 'sum'),
    visit_frequency=('OrderID', 'count')  # Menghitung jumlah transaksi
).reset_index()

# Menampilkan data pelanggan
customer_data.head()
```

```python
# Segmentasi pelanggan berdasarkan total pengeluaran
customer_data['Customer_Segment'] = pd.cut(customer_data['total_spent'],
                                           bins=[0, 1000, 5000, float('inf')],
                                           labels=['Low-value', 'Medium-value', 'High-value'])

# Segmentasi berdasarkan frekuensi kunjungan
customer_data['Visit_Frequency_Segment'] = pd.cut(customer_data['visit_frequency'],
                                                  bins=[0, 5, 15, float('inf')],
                                                  labels=['Infrequent', 'Moderate', 'Frequent'])

# Menampilkan data pelanggan setelah segmentasi
customer_data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menghitung jumlah pelanggan di setiap segmen
customer_segment_counts = customer_data['Customer_Segment'].value_counts()

# Mencari index bar dengan jumlah pelanggan tertinggi
max_customer_segment_index = np.argmax(customer_segment_counts)

# Menyiapkan warna: Semua bar pastel pink kecuali yang tertinggi magenta
colors = ['#FFB6C1'] * len(customer_segment_counts)  # Pastel pink untuk semua bar
colors[max_customer_segment_index] = 'magenta'  # Magenta untuk bar dengan nilai tertinggi

# Membuat bar chart untuk jumlah pelanggan di setiap segmen nilai (Customer Segment)
plt.figure(figsize=(8,6))
plt.bar(customer_segment_counts.index, customer_segment_counts.values, color=colors)
plt.title('Customer Segmentation Based on Total Spending')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Menyimpan visualisasi jika diperlukan
plt.savefig('customer_segment_based_on_spending.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menghitung jumlah pelanggan di setiap segmen frekuensi kunjungan
visit_frequency_counts = customer_data['Visit_Frequency_Segment'].value_counts()

# Mencari index bar dengan jumlah pelanggan tertinggi
max_visit_frequency_index = np.argmax(visit_frequency_counts)

# Menyiapkan warna: Semua bar pastel pink kecuali yang tertinggi magenta
colors = ['#FFB6C1'] * len(visit_frequency_counts)  # Pastel pink untuk semua bar
colors[max_visit_frequency_index] = 'magenta'  # Magenta untuk bar dengan nilai tertinggi

# Membuat bar chart untuk frekuensi kunjungan (Visit Frequency Segment)
plt.figure(figsize=(8,6))
plt.bar(visit_frequency_counts.index, visit_frequency_counts.values, color=colors)
plt.title('Customer Segmentation Based on Visit Frequency')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Menyimpan visualisasi jika diperlukan
plt.savefig('visit_frequency_segment.png')
```

```python
# Menggabungkan data pelanggan dengan segmentasi ke dalam dataset transaksi
df_merged = pd.merge(df, customer_data[['CustID', 'Customer_Segment']], on='CustID', how='left')

# Menghitung total penjualan setiap produk dalam setiap segmen pelanggan
top_products_per_segment = df_merged.groupby(['Customer_Segment', 'ProductSKU']).agg(
    total_sales=('TotalSales', 'sum'),
    total_quantity=('OrderQuantity', 'sum')
).reset_index()

# Mengurutkan produk berdasarkan total sales di setiap segmen
top_products_per_segment = top_products_per_segment.sort_values(by=['Customer_Segment', 'total_sales'], ascending=[True, False])

# Menampilkan top 5 produk teratas di setiap segmen
for segment in top_products_per_segment['Customer_Segment'].unique():
    print(f"\nTop 5 Products for {segment} Customers:")
    display(top_products_per_segment[top_products_per_segment['Customer_Segment'] == segment].head(5))
```

```python
# Melihat tren produk yang dibeli oleh setiap segmen pelanggan
trend_analysis = df_merged.groupby(['Customer_Segment', 'ProductSKU']).agg(
    avg_quantity=('OrderQuantity', 'mean'),
    avg_total_sales=('TotalSales', 'mean')
).reset_index()

# Menampilkan data tren untuk masing-masing segmen
for segment in trend_analysis['Customer_Segment'].unique():
    print(f"\nTrends for {segment} Customers:")
    display(trend_analysis[trend_analysis['Customer_Segment'] == segment].head(5))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat visualisasi untuk produk teratas di setiap segmen
for segment in top_products_per_segment['Customer_Segment'].unique():
    top_5_products = top_products_per_segment[top_products_per_segment['Customer_Segment'] == segment].head(5)

    # Mencari index bar dengan nilai total sales tertinggi
    max_sales_index = np.argmax(top_5_products['total_sales'])

    # Menyiapkan warna: Semua bar pastel pink kecuali yang tertinggi magenta
    colors = ['#FFB6C1'] * len(top_5_products)  # Pastel pink untuk semua bar
    colors[max_sales_index] = 'magenta'  # Magenta untuk bar dengan nilai tertinggi

    # Membuat bar chart untuk produk teratas di segmen saat ini
    plt.figure(figsize=(10, 6))
    plt.bar(top_5_products['ProductSKU'], top_5_products['total_sales'], color=colors)
    plt.title(f"Top 5 Products for {segment} Customers")
    plt.ylabel('Total Sales')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Menyimpan visualisasi untuk setiap segmen
    plt.savefig(f'top_5_products_{segment}_customers.png')
```
